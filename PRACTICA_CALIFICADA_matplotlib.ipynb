EJERCICIOS DE MATPLOTLIB DE 1 AL 5 EN UNO: Desarrollar el diagrama de barras, sectores, histograma, líneas, cajas, con su respectiva
leyenda y ejes según se requiera:  


```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Datos originales
datos = [2, 3, 5, 1, 6, 7, 8, 0, 8, 9, 6, 4, 4, 3, 2, 6, 3, 3, 2, 1, 1, 5, 6, 7, 4, 3, 2, 5, 4, 6]

# Crear tabla de frecuencias con intervalos de amplitud 2
intervalos = pd.interval_range(start=0, end=10, freq=2, closed='left')
tabla_frec = pd.cut(datos, bins=intervalos)
frecuencias = tabla_frec.value_counts().sort_index()
df_frecuencias = pd.DataFrame({
    'Intervalo': [str(i) for i in frecuencias.index],
    'Frecuencia': frecuencias.values
})

# Crear figura con 5 subgráficos
fig, axs = plt.subplots(3, 2, figsize=(12, 10))
fig.delaxes(axs[2, 1])  # Eliminar celda vacía para ajustar

# 1. Diagrama de barras
axs[0, 0].bar(df_frecuencias['Intervalo'], df_frecuencias['Frecuencia'], color='skyblue', edgecolor='black')
axs[0, 0].set_title("Diagrama de Barras")
axs[0, 0].set_xlabel("Intervalos")
axs[0, 0].set_ylabel("Frecuencia")

# 2. Diagrama de sectores
axs[0, 1].pie(df_frecuencias['Frecuencia'], labels=df_frecuencias['Intervalo'], autopct='%1.1f%%', colors=plt.cm.Paired.colors)
axs[0, 1].set_title("Diagrama de Sectores")

# 3. Histograma
axs[1, 0].hist(datos, bins=range(0, 11, 2), color='lightgreen', edgecolor='black')
axs[1, 0].set_title("Histograma")
axs[1, 0].set_xlabel("Intervalos")
axs[1, 0].set_ylabel("Frecuencia")

# 4. Gráfico de líneas
axs[1, 1].plot(df_frecuencias['Intervalo'], df_frecuencias['Frecuencia'], marker='o', linestyle='-', color='red')
axs[1, 1].set_title("Gráfico de Líneas")
axs[1, 1].set_xlabel("Intervalos")
axs[1, 1].set_ylabel("Frecuencia")

# 5. Diagrama de cajas
axs[2, 0].boxplot(datos, patch_artist=True, boxprops=dict(facecolor='orange', color='black'))
axs[2, 0].set_title("Diagrama de Cajas")
axs[2, 0].set_ylabel("Valores")

plt.tight_layout()
plt.show()
```

9. En el estudio estadístico realizado en un instituto se han obtenido los siguientes
datos:
o Peso (kg): [50, 55), [55, 60), [60, 65), [65, 70), [70, 75]
o Número de alumnos: 10, 40, 25, 20, 5
a) Organiza una tabla estadística. b) Construye el histograma y el
polígono de frecuencias.  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Datos ---
intervalos = ["50-54", "55-59", "60-64", "65-69", "70-75"]
frecuencias = np.array([10, 40, 25, 20, 5])

# --- a) Tabla estadística ---
tabla = pd.DataFrame({
    "Intervalo de peso (kg)": intervalos,
    "Frecuencia": frecuencias
})
print("Tabla estadística:\n")
print(tabla.to_string(index=False))

# Calcular puntos medios para el polígono de frecuencias
puntos_medios = np.array([52, 57, 62, 67, 72.5])

# --- b) Histograma ---
plt.figure(figsize=(10,6))
# Para que el histograma se alinee con los intervalos, usamos bins
bins = [50, 55, 60, 65, 70, 75]
plt.hist(
    x=np.repeat(puntos_medios, frecuencias),  # Repetimos los puntos según su frecuencia
    bins=bins,
    color='lightblue',
    edgecolor='black',
    rwidth=0.9
)
plt.xlabel("Peso (kg)")
plt.ylabel("Número de alumnos")
plt.title("Histograma del peso de los alumnos")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Polígono de frecuencias ---
plt.figure(figsize=(10,6))
plt.plot(puntos_medios, frecuencias, marker='o', color='red', linestyle='-')
plt.xlabel("Peso (kg)")
plt.ylabel("Número de alumnos")
plt.title("Polígono de frecuencias del peso de los alumnos")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

b) Dibuja el histograma correspondiente.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# --- Datos ---
temperaturas = np.array([22, 24, 26, 28, 30, 32, 34])

# --- a) Tabla de frecuencias ---
# Definir intervalos de 2 grados: de 22 a 34
bins = np.arange(22, 36, 2)  # 22,24,26,28,30,32,34
labels = [f"{bins[i]}-{bins[i+1]-1}" for i in range(len(bins)-1)]

# Clasificar los datos en los intervalos
categorias = pd.cut(temperaturas, bins=bins, right=False, labels=labels)
frecuencia = categorias.value_counts().sort_index()

# Crear DataFrame de la tabla de frecuencias
tabla = pd.DataFrame({
    "Intervalo (°C)": frecuencia.index,
    "Frecuencia": frecuencia.values
})
print("Tabla de frecuencias:\n")
print(tabla.to_string(index=False))

# --- b) Histograma ---
plt.figure(figsize=(8,6))
plt.hist(temperaturas, bins=bins, color='lightcoral', edgecolor='black', rwidth=0.9)
plt.xlabel("Temperatura (°C)")
plt.ylabel("Frecuencia")
plt.title("Histograma de temperaturas diarias")
plt.xticks(bins)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

15. La distribución del tiempo libre semanal de un grupo de personas es la
siguiente:
o Deporte: 30%
o Lectura: 20%
o Cine: 15%
o Viajes: 10%
o Otras actividades: 25%
a) Representa estos datos mediante un diagrama de sectores.
b) Calcula el ángulo correspondiente a cada sector.

```python
import matplotlib.pyplot as plt

# --- Datos ---
actividades = ["Deporte", "Lectura", "Cine", "Viajes", "Otras actividades"]
porcentajes = [30, 20, 15, 10, 25]  # en %

# --- a) Diagrama de sectores (Pie Chart) ---
plt.figure(figsize=(8,8))
plt.pie(porcentajes, labels=actividades, autopct='%1.1f%%', startangle=90, colors=['#4CAF50','#2196F3','#FF9800','#9C27B0','#FFC107'])
plt.title("Distribución del tiempo libre semanal")
plt.show()

# --- b) Cálculo de los ángulos para cada sector ---
# Ángulo = porcentaje * 360 / 100
angulos = [p * 360 / 100 for p in porcentajes]

print("Ángulo correspondiente a cada sector (en grados):\n")
for actividad, angulo in zip(actividades, angulos):
    print(f"{actividad}: {angulo:.1f}°")
```

```python
import matplotlib.pyplot as plt

# --- Datos ---
transportes = ["Coche", "Bicicleta", "Caminando", "Transporte público"]
porcentajes = [40, 25, 20, 15]  # en %

# --- a) Diagrama de sectores (Pie Chart) ---
plt.figure(figsize=(8,8))
plt.pie(porcentajes, labels=transportes, autopct='%1.1f%%', startangle=90,
        colors=['#FF5722', '#4CAF50', '#2196F3', '#FFC107'])
plt.title("Medios de transporte utilizados por los estudiantes")
plt.show()

# --- b) Cálculo de los ángulos para cada sector ---
# Fórmula: Ángulo = porcentaje * 360 / 100
angulos = [p * 360 / 100 for p in porcentajes]

print("Ángulo correspondiente a cada sector (en grados):\n")
for transporte, angulo in zip(transportes, angulos):
    print(f"{transporte}: {angulo:.1f}°")
```
